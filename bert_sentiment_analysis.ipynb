```python
!pip install -q transformers
!pip install -q torch
from transformers import pipeline
sentiment_analyzer = pipeline("sentiment-analysis")
results = sentiment_analyzer([
    "I love how fast this got resolved!",
    "No one answered my request.",
    "What a disaster. I hate the service!",
    "Fantastic support, I'm impressed!",
    "I'm seriously disappointed and frustrated."
])

for sentence, result in zip([
    "I love how fast this got resolved!",
    "No one answered my request.",
    "What a disaster. I hate the service!",
    "Fantastic support, I'm impressed!",
    "I'm seriously disappointed and frustrated."
], results):
    label = result['label']
    score = round(result['score'], 2)
    emoji = "😊" if label == "POSITIVE" else "😡"
    print(f"{sentence} → {label} ({score}) {emoji}")
```
